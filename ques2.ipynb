```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def derivative(t,y):
    return -y + 40

t=np.arange(0,5,0.001)
result = solve_ivp(derivative, (t[0], t[-1]), [45], t_eval=t)
plt.plot(result.t , result.y[0] ,label='Solution')
plt.legend()
plt.grid()
v = plt.axis()
plt.axis([t[0], t[-1], v[2], v[3]])
plt.show()
```

```python
def derivative(t,ivec):
    return [ivec[1], -ivec[0]+40] 

t=np.arange(0,330,0.001)
result = solve_ivp(derivative, (t[0], t[-1]), [0,45], t_eval=t)
plt.plot(result.t, result.y[0], label='Solution')
plt.legend()
plt.grid()
v = plt.axis()
plt.axis([t[0], t[-1], v[2], v[3]])
plt.show()
```
